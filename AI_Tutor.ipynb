```python
import os
import gradio as gr
from pymongo import MongoClient
from sentence_transformers import SentenceTransformer
from google.colab import userdata
userdata.get('MONGO_URI')
client = MongoClient(userdata.get('MONGO_URI'))
dbName = "langchain_demo"
collectionName = "collection_of_text_blobs"
embeddingCollectionName = "collection_of_text_embeddings"
collection = client[dbName][collectionName]
embeddingCollection = client[dbName][embeddingCollectionName]

model = SentenceTransformer("paraphrase-MiniLM-L6-v2")

with gr.Blocks(theme=gr.themes.Base(), title="File Upload and Embeddings") as main_interface:
    gr.Markdown("## File Upload and Embeddings")
    file_output = gr.Interface(inputs="file", outputs="textbox", fn=upload_file)
    print(file_output)
    def upload_file(file):
        try:
            processed_data = file.read()

            save_path = "/content/flagged/file"
            os.makedirs(save_path, exist_ok=True)
            file_path = os.path.join(save_path, file.name)
            with open(file_path, "wb") as f:
                f.write(processed_data)

            document = {"file_name": file.name, "file_path": file_path, "processed_data": processed_data}
            collection.insert_one(document)
            embeddings = model.encode(processed_data.decode('utf-8'))
            embedding_document = {"file_name": file.name, "embedding": embeddings.tolist()}
            embeddingCollection.insert_one(embedding_document)

            return f"File Uploaded and Processed Successfully! File saved at {file_path}"
        except Exception as e:
            return f"Error Uploading File: {str(e)}"


main_interface.launch()
```
